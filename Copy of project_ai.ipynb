* We support Llama, Mistral, CodeLlama, TinyLlama, Vicuna, Open Hermes etc
* And Yi, Qwen ([llamafied](https://huggingface.co/models?sort=trending&search=qwen+llama)), Deepseek, all Llama, Mistral derived archs.
* We support 16bit LoRA or 4bit QLoRA. Both 2x faster.
* `max_seq_length` can be set to anything, since we do automatic RoPE Scaling via [kaiokendev's](https://kaiokendev.github.io/til) method.
* [**NEW**] With [PR 26037](https://github.com/huggingface/transformers/pull/26037), we support downloading 4bit models **4x faster**! [Our repo](https://huggingface.co/unsloth) has Llama, Mistral 4bit models.

```python
from unsloth import FastLanguageModel
import torch
max_seq_length = 2048 # Choose any! We auto support RoPE Scaling internally!
dtype = None # None for auto detection. Float16 for Tesla T4, V100, Bfloat16 for Ampere+
load_in_4bit = True # Use 4bit quantization to reduce memory usage. Can be False.

# 4bit pre quantized models we support for 4x faster downloading + no OOMs.
fourbit_models = [
    "unsloth/mistral-7b-bnb-4bit",
    "unsloth/mistral-7b-instruct-v0.2-bnb-4bit",
    "unsloth/llama-2-7b-bnb-4bit",
    "unsloth/gemma-7b-bnb-4bit",
    "unsloth/gemma-7b-it-bnb-4bit", # Instruct version of Gemma 7b
    "unsloth/gemma-2b-bnb-4bit",
    "unsloth/gemma-2b-it-bnb-4bit", # Instruct version of Gemma 2b
    "unsloth/llama-3-8b-bnb-4bit", # [NEW] 15 Trillion token Llama-3
] # More models at https://huggingface.co/unsloth

model, tokenizer = FastLanguageModel.from_pretrained(
    model_name = "unsloth/Phi-3-mini-4k-instruct",
    max_seq_length = max_seq_length,
    dtype = dtype,
    load_in_4bit = load_in_4bit,
    # token = "hf_...", # use one if using gated models like meta-llama/Llama-2-7b-hf
)
```

```python
import pdfplumber
import os
import shutil

output_folder = 'extracted_text'

if os.path.exists(output_folder):
    shutil.rmtree(output_folder)
os.makedirs(output_folder)

with pdfplumber.open('learningpython.pdf') as pdf:
    for page_num, page in enumerate(pdf.pages):
        text = page.extract_text()
        filename = f'{output_folder}/page_{page_num + 1}.txt'
        with open(filename, 'w', encoding='utf-8') as file:
            file.write(text or "No text found on this page.")

        if page_num == 50:
            break

print(f'Extracted text from pages 1 to {page_num + 1} into the folder "{output_folder}".')
```

```python
from langchain.document_loaders import DirectoryLoader

DATA_PATH = "extracted_text"

def load_documents():
    loader = DirectoryLoader(DATA_PATH, glob="*.txt")
    documents = loader.load()
    return documents

documents = load_documents()
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter


text_spiltter = RecursiveCharacterTextSplitter(
    chunk_size=1000,
    chunk_overlap=500,
    length_function=len,
    add_start_index=True,
)

chunks = text_spiltter.split_documents(documents)

```

```python
print(f"Split {len(documents)} documents into {len(chunks)} chunks")
```

```python
import os
import shutil
from langchain_chroma import Chroma
from langchain_community.embeddings import HuggingFaceBgeEmbeddings

CHROMA_PATH = "chroma"

if os.path.exists(CHROMA_PATH):
    shutil.rmtree(CHROMA_PATH)

# Set up the BGE embeddings
model_name = "BAAI/bge-small-en"
model_kwargs = {"device": "cpu"}
encode_kwargs = {"normalize_embeddings": True}
hf_embeddings = HuggingFaceBgeEmbeddings(
    model_name=model_name, model_kwargs=model_kwargs, encode_kwargs=encode_kwargs
)

db = Chroma.from_documents(chunks, hf_embeddings, persist_directory=CHROMA_PATH)

print(f"Saved {len(chunks)} chunks to {CHROMA_PATH}")
```

```python
# reading from database


import os
import shutil
from langchain_chroma import Chroma
from langchain_community.embeddings import HuggingFaceBgeEmbeddings

CHROMA_PATH = "chroma"

# Set up the BGE embeddings
model_name = "BAAI/bge-small-en"
model_kwargs = {"device": "cpu"}  # Use "cuda" for GPU if available
encode_kwargs = {"normalize_embeddings": True}

# Initialize the HuggingFace BGE Embeddings
bge_embeddings = HuggingFaceBgeEmbeddings(
    model_name=model_name, model_kwargs=model_kwargs, encode_kwargs=encode_kwargs
)

# Initialize the Chroma database with BGE embeddings
db = Chroma(persist_directory=CHROMA_PATH, embedding_function=bge_embeddings)

# # Assuming 'chunks' is a list of texts/documents you want to store in the database
# db.index_documents(chunks)

```

```python
query_text = "who is author of learning python??"
results = db.similarity_search_with_relevance_scores(query_text, k=3)
```

```python
if len(results) == 0 or results[0][1] < 0.5:
  print(f"Unable to find matching")
else:
  context_text = "\n\n---\n\n".join([doc.page_content for doc, _score in results])
  print(context_text)
```

```python
# alpaca_prompt = Copied from above
from transformers import TextStreamer

# FastLanguageModel.for_inference(model) # Enable native 2x faster inference

alpaca_prompt = """Below is an instruction that describes a task, paired with an input that provides further context. Write a response that appropriately completes the request.

### Instruction:
{}

### Input:
{}

### Response:
{}"""

inputs = tokenizer(
[
    alpaca_prompt.format(
        query_text, # instruction
        context_text, # input
        "", # output - leave this blank for generation!
    )
], return_tensors = "pt").to("cuda")

text_streamer = TextStreamer(tokenizer)
_ = model.generate(**inputs, streamer = text_streamer, max_new_tokens = 128)
```

```python
streamlit_script = """
import streamlit as st
from langchain_chroma import Chroma
from langchain_community.embeddings import HuggingFaceBgeEmbeddings

# Initialize embeddings and database
model_name = "BAAI/bge-small-en"
model_kwargs = {"device": "cpu"}
encode_kwargs = {"normalize_embeddings": True}
bge_embeddings = HuggingFaceBgeEmbeddings(model_name=model_name, model_kwargs=model_kwargs, encode_kwargs=encode_kwargs)
db = Chroma(persist_directory="chroma", embedding_function=bge_embeddings)

def query_database(query):
    results = db.similarity_search_with_relevance_scores(query, k=3)
    if len(results) == 0 or results[0][1] < 0.5:
        return "No relevant results found."
    else:
        return "\\n\\n---\\n\\n".join([doc.page_content for doc, _ in results])

# Streamlit UI components
st.title('Text-Based Query Application')
query = st.text_input('Enter your query here:')
if st.button('Search'):
    if query:
        response = query_database(query)
        st.text_area("Response:", response, height=300)
    else:
        st.write("Please enter a query to search.")
"""

# Save the Streamlit script to a file
with open('streamlit_app.py', 'w') as file:
    file.write(streamlit_script)
```
